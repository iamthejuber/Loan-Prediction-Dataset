```python
def empirical_analysis(col):
    mu = train_df[col].mean()
    sigma = train_df[col].std()
    minus1 = mu - sigma ; plus1 = mu + sigma
    minus2 = mu - (2*sigma) ; plus2 = mu + (2*sigma)
    minus3 = mu - (3*sigma) ; plus3 = mu + (3*sigma)
    x1 = train_df[(train_df[col] > minus1) & (train_df[col] < plus1)]
    x2 = train_df[(train_df[col] > minus2) & (train_df[col] < plus2)]
    x3 = train_df[(train_df[col] > minus3) & (train_df[col] < plus3)]
    return(x1,x2,x3)
```

$np.where$

- for binary conditions

- If any condition is true will modify

- otherwise dont modify

- np.where(condition,modify_value,same_value)

- np.where(condition,True,False)

```python
con1=train_df['ApplicantIncome']>ub
con2=train_df['ApplicantIncome']<lb
con1|con2
median=train_df['ApplicantIncome'].median()
#np.where(<condition>,<modified_vale>,<same>)

train_df['ApplicantIncome']=np.where(con1|con2,
                                    median,
                                    train_df['ApplicantIncome'])

train_df['imputed_column']=np.where(con1|con2,
                                    median,
                                    train_df['ApplicantIncome'])
train_df
```
